```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



df = pd.read_csv('data 2.csv', encoding='latin1')
x = df['no2'].dropna()

ar = 0.05 * (1023036661 % 7)
br = 0.3 * ((102303661 % 5) + 1)
z = x + ar * np.sin(br * x)

mean = np.mean(z)
var = np.var(z)
lam = 1 / (2 * var)
c = np.sqrt(lam / np.pi)


print(f"Lambda: {lam}")
print(f"Mu:     {mean}")
print(f"c:      {c}")
```
